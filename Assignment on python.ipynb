```python
a= input("Enter the value of a here:")
b= input("Enter the value of b here:")
sum_result= float(a)+float(b)
difference= float(a)-float(b)
product= float(a)*float(b)
quotient= float(a)/float(b)
power= int(a)**int(b)
print("The sum of a and b is",sum_result)
print("The difference of a and b is", difference)
print("The product of a and b is",product)
print("The quotient of a and b is", quotient)
print("The power of a and b is", power)
```
